```python
def encodePT(teams):
    value = team_dic.get(teams)
    return value
PossessionTeam_encoded = train['PossessionTeam'].apply(encodePT)
HomeTeamAbbr_encoded = train['HomeTeamAbbr'].apply(encodePT)
VisitorTeamAbbr_encoded = train['VisitorTeamAbbr'].apply(encodePT)

def encodeTeam(teams):
    if (teams == 'home'):
        return 0
    elif (teams == 'away'):
        return 1
Team_encoded = train['Team'].apply(encodeTeam)

def encodePD(teams):
    if (teams == 'left'):
        return 0
    elif (teams == 'right'):
        return 1
PlayDirection_encoded = train['PlayDirection'].apply(encodePD)

def encodePosition(teams):
    value = Position_dict.get(teams)
    return value
Position_encoded = train['Position'].apply(encodePosition)

def encodeDefensePersonnel(teams):
    value = DefensePersonnel_dict.get(teams)
    if value == None:
        return 42
    return value
DefensePersonnel_encoded = train['DefensePersonnel'].apply(encodeDefensePersonnel)

#def encodeTurf(teams):
#    value = Turf_dict.get(teams)
#    return value
#Turf_encoded = train['Turf'].apply(encodeTurf)

def encodeOF(teams):
    value = OffenseFormation_dict.get(teams)
    return value
OffenseFormation_encoded = train['OffenseFormation'].apply(encodeOF)
```

```python
#To get the number of WideReceivers
OffensePersonnel_split =  train.OffensePersonnel.str.split(",")
def extractWR(personnel):
    WR = personnel[-1]
    WRSplit = WR.split()
    numberOfWR = WRSplit[0]
    return numberOfWR
widereceivers = OffensePersonnel_split.apply(extractWR)

def extractTE(personnel):
    TE = personnel[-2]
    TESplit = TE.split()
    numberOfTE = TESplit[0]
    return numberOfTE
TightEnds = OffensePersonnel_split.apply(extractTE)

df = pd.DataFrame()
def calculate_age(born):
    born = datetime.strptime(born, "%m/%d/%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

df['age'] = train['PlayerBirthDate'].apply(calculate_age)

bins_age = [19, 25, 28, 31, 34, 37, 45]
Age_Categorized = pd.cut(df['age'], bins= bins_age, labels = [1,2,3,4,5,6])

def convertInches(height):
    height = height.split('-')
    height_in = int(height[0])*12 + int(height[1])
    return height_in
playerHeight = train.PlayerHeight.apply(convertInches)

height_sq = playerHeight.pow(2)
BMI = train.PlayerWeight.div(height_sq)
BMI = BMI.mul(703)
```

```python
#Finding x and y components of Velocity(Speed)

#To calculate the x and y components of Speed, we need to normalize Orientation first
def normalize_Orientation(x):
    if x<=90:
        x = 90 -x
    elif x<=180:
        x = x -90
    elif x<=270:
        x = 270 -x
    elif x<=360:
        x = x -270
    return x

def getCos(orientation):
    angle = (orientation*(math.pi))/180
    return math.cos(angle)

def getSin(orientation):
    angle = (orientation*(math.pi))/180
    return math.sin(angle)

def getSpeed(speed):
    return speed

Orientation_normalized = train.Orientation.apply(normalize_Orientation)
train = train.drop(['Orientation'], axis =1)
train['Orientation'] = Orientation_normalized
#train['Orientation'].describe()
x_velocity = train.Orientation.apply(getCos).multiply(train.S)
y_velocity = train.Orientation.apply(getSin).multiply(train.S)
```
